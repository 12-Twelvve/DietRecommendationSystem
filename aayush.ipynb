```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numerical_cols = ['Calories', 'Fats', 'Proteins', 'Iron', 'Calcium','Sodium', 'Potassium', 'Carbohydrates', 'Fibre', 'VitaminD', 'Sugars']
```

```python
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (15,12))
ax = fig.gca()
data.select_dtypes(include=np.number).hist(ax =ax)
plt.show()
```

```python
data[data['cluster']==4]
```

```python
# Step 2: Prepare the input features and target variable
X = df[numerical_cols]
y = df['Food_items']
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
```

```python
data['cluster'] = kmeans.labels_
```

```python
data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_rf = data.drop(['Food_items'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_rf, y, test_size=0.2, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

# Make predictions on the test set
y_pred = rf.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
